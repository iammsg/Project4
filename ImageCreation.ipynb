```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt

image_center = ndimage.imread('Images/center_view.jpg')
image_left = ndimage.imread('Images/left_view.jpg')
image_right= ndimage.imread('Images/right_view.jpg')
```

```python
fig = plt.figure(1, figsize=(20,20))
ax=plt.subplot(1, 3,2)
plt.imshow(image_center)
ax.set_title("Center Camera")
ax.axis('off')
ax=plt.subplot(1, 3,1)
plt.imshow(image_left)
ax.set_title("Left Camera")
ax.axis('off')
ax=plt.subplot(1, 3,3)
plt.imshow(image_right)
ax.set_title("Right Camera")
ax.axis('off')
plt.savefig('Images/Combined_Camera.png',bbox_inches='tight')
```

```python
image_center_flipped=np.fliplr(image_center)

fig = plt.figure(1, figsize=(20,20))
ax=plt.subplot(1, 2,1)
plt.imshow(image_center)
ax.set_title("Original Image")
ax.axis('off')
ax=plt.subplot(1, 2,2)
plt.imshow(image_center_flipped)
ax.set_title("Image Flipped")
ax.axis('off')
plt.savefig('Images/Flipped_Images.png',bbox_inches='tight')
```
